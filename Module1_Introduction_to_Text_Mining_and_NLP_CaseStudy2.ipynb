```python

# 9.Count the number ofarticles in file ‘Macbeth.txt’(Articlesare ‘a’, ‘an’&’the’ )and save the file as ‘Macbeth-ArticlesRemoved.txt’

import re

# Read Macbeth.txt
with open("Macbeth.txt", "r", encoding="utf-8") as file:
    text = file.read()

# Find articles (a, an, the) - case insensitive
articles = re.findall(r'\b(a|an|the)\b', text, flags=re.IGNORECASE)

# Count articles
article_count = len(articles)

# Remove articles
cleaned_text = re.sub(r'\b(a|an|the)\b', '', text, flags=re.IGNORECASE)

# Save cleaned content to new file
with open("Macbeth-ArticlesRemoved.txt", "w", encoding="utf-8") as file:
    file.write(cleaned_text)

print("Number of articles found:", article_count)
print("File saved as Macbeth-ArticlesRemoved.txt")



```

```python
# 10.Remove allthe punctuations from the file. (Punctuations are .,’ : ;# ( ) etc.)and save as ‘Macbeth-punctuationsRemoved.txt’

import string

# Read Macbeth.txt
with open("Macbeth.txt", "r", encoding="utf-8") as file:
    text = file.read()

# Remove all punctuations
no_punctuation_text = text.translate(
    str.maketrans('', '', string.punctuation)
)

# Save cleaned text
with open("Macbeth-punctuationsRemoved.txt", "w", encoding="utf-8") as file:
    file.write(no_punctuation_text)

print("Punctuations removed successfully.")
print("File saved as Macbeth-punctuationsRemoved.txt")
```

```python

#13.For each alphabet print the frequency of names present in ‘names’ corpora
import nltk
from nltk.corpus import names

# Get all names
all_names = names.words()

# Dictionary to store frequency
alphabet_count = {}

# Count names starting with each alphabet
for name in all_names:
    first_letter = name[0].upper()
    alphabet_count[first_letter] = alphabet_count.get(first_letter, 0) + 1

# Print frequency
for letter in sorted(alphabet_count):
    print(letter, ":", alphabet_count[letter])
    
```

```python
#14.Plot a bar graph showing frequency of names for each alphabet

import nltk
from nltk.corpus import names
import matplotlib.pyplot as plt


all_names = names.words()
alphabet_freq = {}

for name in all_names:
    letter = name[0].upper()
    alphabet_freq[letter] = alphabet_freq.get(letter, 0) + 1

# Sort alphabets
alphabets = sorted(alphabet_freq.keys())
frequencies = [alphabet_freq[a] for a in alphabets]

# Plot bar graph
plt.figure()
plt.bar(alphabets, frequencies)
plt.xlabel("Alphabets")
plt.ylabel("Number of Names")
plt.title("Frequency of Names for Each Alphabet (NLTK Names Corpus)")
plt.show()
```

```python

#15.Add appropriate titles, X-axis labels and Y-axis labels to the above created bar graph


# Plot bar graph
plt.bar(alphabets, frequencies)

# ✅ Adding labels and title
plt.title("Frequency of Names by Alphabet (NLTK Names Corpus)")
plt.xlabel("Alphabets (A–Z)")
plt.ylabel("Number of Names")

# Show graph
plt.show()
```
