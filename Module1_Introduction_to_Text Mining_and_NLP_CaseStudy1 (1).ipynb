# 3. You are given a docx file named “NLP.docx”. Perform the following tasks:
    •Use DOCX module to open the file and perform the following tasks:
    •How many paragraphs are there in the file?
    •For each paragraph, print the number of words it contains
    •Read the content from “Welcome.txt” file which you created in question number 2 and make that content the first paragraph in “NLP.docx” file

```python
!pip install python-docx
```

```python
#How many paragraphs are there in the file?

from docx import Document

doc = Document("NLP.docx")
print("How many paragraphs are there in the file :: ", len(doc.paragraphs))
```

```python
#For each paragraph, print the number of words it contains

from docx import Document
doc = Document("NLP.docx")

count =1
for i in doc.paragraphs:
    word = i.text.split()
    print("Paragraph ", count, "has ",len(word) )
    count +=1
```

```python
#Read the content from “Welcome.txt” file which you created in question number 2 and make that content the first paragraph in “NLP.docx” file

from pathlib import Path
from docx import Document

txt_doc = Path("Welcome.txt")
content = txt_doc.read_text()
document = Document("NLP.docx")
document.paragraphs[0].insert_paragraph_before(content)
document.save("NLP.docx")
```

# 4. You are given a csv file named “EmployeeDetails.csv”. Perform the following tasks:
    •Use pandas module to read the csv file
    •Split the name as first name and last name and make two separate columnsas shown
    •Delete “Name” Column 
    •Increment the “Salary” Column by 10% of the existing valueand save changes in same column onlyas shown 
    •Save changes to a new file named “Employee_Data.csv”

```python
#•Use pandas module to read the csv file
#•Split the name as first name and last name and make two separate columnsas shown

import pandas as pd

csv_file = pd.read_csv("EmployeeDetails.csv")

csv_file.columns = csv_file.columns.str.strip()

csv_file["First Name"] = csv_file["Name"].str.split().str[0]
csv_file["Last Name"] = csv_file["Name"].str.split().str[-1]

csv_file.head()

```

```python
#Delete “Name” Column 

csv_file = csv_file.drop(columns=["Name"])
csv_file.head()

```

```python
#Increment the “Salary” Column by 10% of the existing value and save changes in same column only as shown

csv_file["Salary"] = csv_file["Salary"]*1.10
csv_file.head()
```

```python
#Save changes to a new file named “Employee_Data.csv”

csv_file.to_csv("Employee_Data.csv", index=False)
```
